--- Day 5: Hydrothermal Venture ---
You come across a field of hydrothermal vents on the ocean floor! These vents constantly produce large, opaque clouds, so it would be best to avoid them if possible.

They tend to form in lines; the submarine helpfully produces a list of nearby lines of vents (your puzzle input) for you to review. For example:

0,9 -> 5,9

8,0 -> 0,8

9,4 -> 3,4

2,2 -> 2,1

7,0 -> 7,4

6,4 -> 2,0

0,9 -> 2,9

3,4 -> 1,4

0,0 -> 8,8

5,5 -> 8,2

Each line of vents is given as a line segment in the format x1,y1 -> x2,y2 where x1,y1 are the coordinates of one end the line segment and x2,y2 are the coordinates of the other end. These line segments include the points at both ends. In other words:

An entry like 1,1 -> 1,3 covers points 1,1, 1,2, and 1,3.

An entry like 9,7 -> 7,7 covers points 9,7, 8,7, and 7,7.

For now, only consider horizontal and vertical lines: lines where either x1 = x2 or y1 = y2.

So, the horizontal and vertical lines from the above list would produce the following diagram:

.......1..

..1....1..

..1....1..

.......1..

.112111211

..........

..........

..........

..........

222111....

In this diagram, the top left corner is 0,0 and the bottom right corner is 9,9. Each position is shown as the number of lines which cover that point or . if no line covers that point. The top-left pair of 1s, for example, comes from 2,2 -> 2,1; the very bottom row is formed by the overlapping lines 0,9 -> 5,9 and 0,9 -> 2,9.

To avoid the most dangerous areas, you need to determine the number of points where at least two lines overlap. In the above example, this is anywhere in the diagram with a 2 or larger - a total of 5 points.

Consider only horizontal and vertical lines. At how many points do at least two lines overlap?

```python
import numpy as np

with open('day5_puzzle_input.txt') as fp:
    line = fp.readline().strip()
    array_list = []
    while line:
        string = '' # empty string to hold characters from each line
        tmp_list = []   # empty list to store individual values
        for i in range(len(line)):
            if 47<ord(line[i])<58:  # if the character is numeric, store it in the string
                string = string + line[i]
            else:
                if string != '':    # if the string is not empty, convert it to an integer and store it in the list.
                    tmp_list.append(int(string))
                    string = ''
        else:
            tmp_list.append(int(string))
        
        arr = np.array(tmp_list)
        arr2 = arr.reshape(2,2)
        if arr2[0,0] == arr2[1,0] or arr2[0,1] == arr2[1,1]:
            array_list.append(arr2[0].tolist())
            array_list.append(arr2[1].tolist())
        line = fp.readline().strip()

x = 0   # value of horizontal range
y = 0   # value of vertical range
for i in range(len(array_list)):
    if array_list[i][1] > y:
        y = array_list[i][1]
    if array_list[i][0] > x:
        x = array_list[i][0]

sonar = np.zeros((y+1,x+1)) # zero matrix to represent the sonar 


for i in range(len(array_list)-1):
    x1 = array_list[i][0]   # value for first horizontal position
    x2 = array_list[i+1][0] # value for second horizontal position
    y1 = array_list[i][1]   # value for the first vertical position
    y2 = array_list[i+1][1] # value for the second horizontal position
    if x1 == x2:
        increase_range = abs(y1 - y2) + 1
        if y1 < y2:
            lower_range = y1
            upper_range = lower_range + increase_range
            sonar[lower_range:upper_range,x1] +=1
        else:
            lower_range = y2
            upper_range = lower_range + increase_range
            sonar[lower_range:upper_range,x1] +=1
    elif y1 == y2:
        increase_range = abs(x1 - x2) + 1
        if x1 < x2:
            lower_range = x1
            upper_range = lower_range + increase_range
            sonar[y1,lower_range:upper_range] +=1
        else:
            lower_range = x2
            upper_range = lower_range + increase_range
            sonar[y1,lower_range:upper_range] +=1

danger_zone = 0 # variable to store how many areas in sonar have a value of >= 2
for x in np.nditer(sonar):  # iterates over each element in array
    if x >= 2:
        danger_zone +=1

print(danger_zone)
```
